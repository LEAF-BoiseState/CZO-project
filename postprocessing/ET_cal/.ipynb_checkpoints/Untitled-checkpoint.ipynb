```python
import matplotlib.pyplot as plt
import pfb_functions as pfn
import numpy as np
import pandas as pd
```

```python
#-----------------------------------------------------------------------------
# Set up run info
#-----------------------------------------------------------------------------
starttime =  1
stoptime  =  6

figure_output_frequency=1

# ,'pfclm1_base40'
case_namelist = ['pfclm1_base1',
                 'pfclm2_base1','pfclm2_base40']
num_case = len(case_namelist)
data_length = stoptime
total_ET_storage_matrix =np.zeros((data_length,num_case),dtype=float)
surface_ET_depth_matrix =np.zeros((data_length,num_case),dtype=float)
```

```python
#### read pfb file
def ET_compute (file_path,case_index,time_step):

    if time_step >0:
        fnameevap = ('%s/%s.out.qflx_evap_tot.%05d.pfb' %(file_path,case_index,time_step))
        fnametran = ('%s/%s.out.qflx_tran_veg.%05d.pfb' %(file_path,case_index,time_step))
        print (fnameevap,fnametran)

        evap_mms = pfn.read(fnameevap)
        tran_mms = pfn.read(fnametran)
        print (evap_mms,tran_mms)

        tran_mh =tran_mms *3600 /1000
        evap_mh =evap_mms *3600 /1000

        surface_tran_volume =evap_mh *30 *30
        surface_evap_volume =evap_mh *30 *30

        total_tran_storage =surface_tran_volume.sum()
        total_evap_storage =surface_evap_volume.sum()

        surface_tran_depth =total_tran_storage /198 /198 /30 /30
        surface_evap_depth =total_evap_storage /198 /198 /30 /30

        total_ET_storage = total_tran_storage + total_evap_storage
        surface_ET_depth = surface_tran_depth + surface_evap_depth

        total_ET_storage_matrix[time_step,case_namelist.index(case_index)]=total_ET_storage
        surface_ET_depth_matrix[time_step,case_namelist.index(case_index)]= surface_ET_depth

    return total_ET_storage_matrix, surface_ET_depth_matrix
```

```python
def write_1D (outputfile_depth_name,outputfile_volume_name,total_ET_storage,surface_ET_depth):
    np.savetxt(outputfile_depth_name,surface_ET_depth,delimiter=",")
    np.savetxt(outputfile_volume_name,total_ET_storage,delimiter=",")
```

```python
def plot_ws_2D (time_step,figure_output_frequency,total_ET_depth,total_ET_volume,sfpath,case_name):
    if time_step%figure_output_frequency==0:
        plt.figure(figsize=(10, 7))
        plt.pcolor(total_ET_depth,cmap='jet')
        plt.colorbar()
        plt.savefig('%s/%s_ET_depth_%05d.png'% (sfpath,case_name,time_step))
        plt.close()     
    
        plt.figure(figsize=(10, 7))
        plt.pcolor(total_ET_volume,cmap='jet')
        plt.colorbar()
        plt.savefig('%s/%s_ET_volume_%05d.png'% (sfpath,case_name,time_step))
        plt.close()
```
